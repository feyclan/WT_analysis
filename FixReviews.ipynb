## Load dataframes and drop books without a title

```python
import pandas as pd
```

```python
books = pd.read_csv("data_sets/duplicate_detection_with_ngrams.csv")
```

```python
books.dropna(subset="Title", inplace=True)
```

```python
reviews = pd.read_csv("data_sets/Reviews_Raw.csv", usecols = ['Title', 'User_id', 'review/score', 'review/time'])
```

```python
reviews.dropna(subset="Title", inplace=True)
```

## Merge columns

```python
merged_df = reviews.merge(books, on='Title', how='left')
merged_df.head()
```

```python
idx = merged_df.groupby(['Title', 'User_id'])['review/time'].idxmax()
```

```python
kept_df = merged_df.loc[idx].reset_index(drop=True)

# Identify the rows that are in the original dataframe but not in the kept rows dataframe
# Here we use the indices in `idx` to compare
original_indices = merged_df.index
kept_indices = idx
removed_indices = original_indices.difference(kept_indices)

# Create the dataframe with the rows that have been removed
removed_df = merged_df.loc[removed_indices].reset_index(drop=True)
```

## Take most recent reviews only, if time is the same, take average of those entries

```python
max_time_df = merged_df.groupby(['Title', 'User_id'])['review/time'].max().reset_index()

# Merge to get the rows with the max review/time
merged_with_max_time = pd.merge(merged_df, max_time_df, on=['Title', 'User_id', 'review/time'])

# Now, group again to calculate the average review/score for the rows with the max review/time
final_df = merged_with_max_time.groupby(['Title', 'User_id', 'review/time']).agg({
    'review/score': 'mean',
    'book_id': 'first',  # 'book_id' should be the same for the same title
}).reset_index()

# Optionally, you can re-arrange the columns to match the original order if needed
final_df = final_df[['Title', 'User_id', 'review/score', 'review/time', 'book_id']]

# To see the rows that were removed, you can find the difference between the original and final_df
removed_df = pd.concat([merged_df, final_df]).drop_duplicates(keep=False).reset_index(drop=True)
```
